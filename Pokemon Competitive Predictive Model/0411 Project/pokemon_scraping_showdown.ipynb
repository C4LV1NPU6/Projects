```python
def web_driver():
    options = webdriver.ChromeOptions()
    options.add_argument("--verbose")
    options.add_argument('--no-sandbox')
    options.add_argument('--headless')
    options.add_argument('--disable-gpu')
    options.add_argument("--window-size=1920, 1200")
    options.add_argument('--disable-dev-shm-usage')
    driver = webdriver.Chrome(options=options)
    return driver
```

```python
driver1 = web_driver()
url = 'https://replay.pokemonshowdown.com/?format=%5BGen%209%5D%20VGC%202024%20Reg%20F&page='
link_tot = []
for i in range(100, 0, -1):
  print(i)
  curr_url = url+str(i)
  driver1.get(curr_url)
  time.sleep(1)
  containers = driver1.find_elements(By.XPATH, '//li')
  link_list = []
  for container in containers:
    link = container.find_element(By.XPATH, './/a').get_attribute('href')
    link_list.append(link)
  link_tot.extend(link_list[6:])
```

```python
print(len(link_tot))
```

```python
list_dic = []
for j, link in enumerate(link_tot):
  new_link = link + '.log'
  driver2 = web_driver()
  driver2.get(new_link)
  time.sleep(3)
  container = driver2.find_element(By.XPATH, './/body')
  lines=container.text.split('\n')
  features = {'team1':[], 'team2':[], 'outcome':0}
  p0 = lines[0][4:]
  p1 = lines[1][4:]
  for i in range(14, 26):
    if lines[i][0:8] == '|poke|p1':
      pokemon = lines[i].split('|')[-2].split(',')[0]
      features['team1'].append(pokemon)
    elif lines[i][0:8] == '|poke|p2':
      pokemon = lines[i].split('|')[-2].split(',')[0]
      features['team2'].append(pokemon)
  for i in lines:
    if i[0:5] == '|win|':
      if i[5:].strip() == p1:
        features['outcome'] = 1
  list_dic.append(features)
  print(j)
```

```python
import csv
csv_columns = ['team1','team2','outcome']
dict_data = list_dic
csv_file = "scraped_data.csv"
try:
    with open(csv_file, 'w') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
        writer.writeheader()
        for data in dict_data:
            writer.writerow(data)
except IOError:
    print("I/O error")
```
